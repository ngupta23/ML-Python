```python
import matplotlib.pyplot as plt
```

```python
# allows the graphs to be shown within notebook
%matplotlib inline 
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

** There are 2 ways to plot - Functional & Object Oriented Method
OO is generally the better way **

** Functional Method **

```python

plt.plot(x,y,'r-') # can add color with r- just as Matlab
# plt.show() # to show plot if using this outside Jupyter Notebook 
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Multiplots
plt.subplot(1,2,1) # similar to Matlab
plt.plot(x,y,'r')

plt.subplot(1,2,2) # similar to Matlab
plt.plot(y,x,'b')
```

** Objected Oriented Method (much better way of using this) **

```python
fig = plt.figure() # created a figure object

# Takes a list of 4 arguments
# Left Bottom Width and Height 
# each value ranges from <0:1>
axes = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes2.set_title('Smaller Plot')
axes1.set_title('Larger Plot')
```

```python
# Below is an example of tuple unpacking
fig,axes = plt.subplots(nrows = 1, ncols = 2)
# This is nothing but an exes manager on top of plt.figure()
# If you use this without any row or col arguments,
# it is same as plt.figure

# Axes is an array of axes
axes[0].plot(x,y)
axes[0].set_title('Plot 1')

axes[1].plot(y,x)
axes[1].set_title('Plot 2')
```

** Figure Size and DPI **

```python
fig = plt.figure(figsize=(4,2), dpi = 100)
# generally playing with dpi is not needed
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize=(8,4))
axes[0].plot(x,y, label='X Squared') # Label needed for Legends
axes[0].plot(x,x**3, label='X Cubed')
axes[1].plot(y,x)
plt.tight_layout() # For fitting tight plots

axes[0].set_title('X2 and X3 v/s X')
axes[0].set_xlabel('X',)
axes[0].set_ylabel('Y')
axes[0].legend(loc=0) # Plots it on the best location on plot

axes[1].set_title('X v/s X Squared')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

### Customizing the plot 

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='red', lw=3,alpha=0.5,ls='-.'
          ,marker='o',markersize=20,markerfacecolor='yellow'
         , markeredgewidth=3, markeredgecolor='blue')
axes.set_xlim([0,2])
axes.set_ylim([0,4])
## 'color'can also specify custom colors in RGB codes
## 'lw' is the line width
## 'alpha' to set the transparency of line
## 'ls' or linestyle': '--' | '-.' | ':' | 'steps'
## 'marker': 'o' | 'x' | '+' | '*'
```
