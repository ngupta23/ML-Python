```python
# Built on top of Matplotlib
# Can be used to call visulaizations from a dataframe directly
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('df1',index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

```python
# Method 1
df1['A'].hist(bins=20)
```

```python
# Method 2
df1['A'].plot(kind='hist',bins=20)
```

```python
# Method 3
# This is what we will use
df1['A'].plot.hist()
```

```python
df2.plot.area(alpha=0.5) # alpha sets the transparency
```

```python
# index needs to be categorical and small
df2.plot.bar(stacked=True)
```

```python
# Line plot
# You can also add Matplotlib arhuments
df1.plot.line(x=df1.index,y='B',figsize=(12,6))
```

```python
df1.plot.scatter(x='A',y='B',c='D') 
# c implies color and that is based on a 3rd column
# this is a 3D plot of sorts
```

```python
df1.plot.scatter(x='A',y='B',c='D',cmap='coolwarm') 
```

```python
# instead of colors, use size for third dimention
df1.plot.scatter(x='A',y='B',s=df1['D']*20) 
```

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
df.head()
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm')
```

```python
df2['a'].plot.kde()
```
